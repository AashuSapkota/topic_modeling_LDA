```python
import json
import pandas as pd
import numpy as np
import re
from datetime import datetime
from gensim import corpora, models
import gensim
import pyLDAvis.gensim_models
import matplotlib.pyplot as plt
```

```python
# --------------------- 1. DATA LOADING ---------------------
with open('onlinekhabar_scraped_articles.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)
```

```python
# Nepali to English digit mapping
digit_map = str.maketrans('०१२३४५६७८९', '0123456789')

def nepali_to_english(num_str):
    return num_str.translate(digit_map)
```

```python
def convert_nepali_date(nepali_date_str):
    months = {
        'बैशाख': 1, 'जेठ': 2, 'असार': 3, 'साउन': 4, 
        'भदौ': 5, 'असोज': 6, 'कात्तिक': 7, 'मंसिर': 8,
        'पुष': 9, 'माघ': 10, 'फाल्गुन': 11, 'चैत': 12
    }

    if nepali_date_str.strip() == "" or "no timestamp found" in nepali_date_str.lower():
        return pd.NaT
    
    parts = nepali_date_str.split()
    
    # Convert year/day to English digits
    year = int(nepali_to_english(parts[0]))
    day = int(nepali_to_english(parts[2]))
    month = months.get(parts[1], 1)

    return datetime(year, month, day)

# Apply the date conversion
df['date'] = df['timestamp'].apply(convert_nepali_date)
df = df.dropna(subset=['date']).sort_values('date').reset_index(drop=True)
```

```python
# --------------------- 3. COMPREHENSIVE NEPALI TEXT PREPROCESSING ---------------------

class NepaliTextPreprocessor:
    def __init__(self):
        # Extensive Nepali Stop Words List
        self.stop_words = set([
            # Pronouns
            "म", "हामी", "तिमी", "तपाईं", "तपाई", "उनी", "उ", "उन", "यो", "त्यो", "यी", "ती", "के", "कुन", "कसको", "कसले",
            # Particles & Postpositions
            "ले", "को", "का", "मा", "बाट", "लाई", "सम्म", "देखि", "गरेर", "भएर", "सँग", "विरुद्ध", "बिच", "तल", "मुनि", "पछि", "अगाडि",
            # Common verbs (conjugated forms)
            "छ", "छन्", "छौ", "छैन", "छैनन्", "हो", "हुन्", "हुन्छ", "हुन्न", "थियो", "थिए", "भयो", "भए", "गर्नु", "गर्छ", "गर्छन्", "गरे", "गर्यो", "गर्न", "गरेको", "गरिएको", "हुने", "भन्नु", "भन्छ", "भन्यो", "रहेछ",
            # Conjunctions
            "र", "तर", "वा", "कि", "यदि", "यद्यपि", "किनभने", "तापनि", "जब", "भने",
            # Adverbs
            "धेरै", "अलि", "एकदम", "यति", "त्यति", "अहिले", "पहिले", "पछि", "सधै", "लगभग", "कहिल्यै", "अझै", "चाँही",
            # Question Words
            "किन", "कहाँ", "कसरी", "कति",
            # Others (Common fillers)
            "भने", "चाहिँ", "पनि", "त", "नै", "ताँ", "होइन", "जस्तो", "जस्तै", "सन्दर्भ", "अनुसार", "बारे", "विषय", "लागि", "गरी", "पर्दै", "जस्तो", "बाहेक", "मध्ये", "द्वारा", "सम्बन्धि", "भर", "भित्र", "बाहिर",
            # Numbers (common)
            "एक", "दुई", "तिन", "चार", "पाँच", "छ", "सात", "आठ", "नौ", "दस", "करोड", "लाख", "हजार"
        ])

        # Suffix patterns ordered by length (longest first)
        self.suffix_patterns = [
            # Verb suffixes (longest first)
            (r'दैछन्$', ''),    # Present continuous plural
            (r'दैछ$', ''),      # Present continuous
            (r'िरहेको$', ''),   # Present continuous (alternative)
            (r'िरहेका$', ''),   # Present continuous plural (alt)
            (r'ाउँछन्$', ''),   # Future habitual plural
            (r'ाउँछ$', ''),     # Future habitual
            (r'्नेछन्$', '्न'),  # Future definite plural
            (r'्नेछ$', '्न'),    # Future definite
            (r'िएको$', ''),     # Perfect participle
            (r'एको$', ''),      # Past participle
            (r'इयो$', ''),      # Past tense
            (r'ियो$', ''),      # Past tense (alt)
            (r'छन्$', ''),      # Present plural
            (r'छ$', ''),        # Present singular
            (r'यो$', ''),       # Past tense (short)
            (r'ने$', ''),       # Future/infinitive
            (r'दै$', ''),       # Continuative
            (r'होस्$', ''),     # Imperative

            # Noun suffixes
            (r'हरूलाई$', ''), # Plural dative
            (r'हरुले$', ''),  # Plural ergative
            (r'हरू$', ''),     # Plural
            (r'हरु$', ''),     # Plural (alt spelling)
            (r'ता$', ''),      # Abstract noun
            (r'ई$', ''),       # Feminine/adjective
            (r'ए$', ''),       # Adjective
            (r'ी$', ''),       # Feminine
            (r'ौ$', ''),       # Feminine honorific
            (r'ड़$', ''),      # Truncation
        ]

        # Irregular word forms
        self.irregular_forms = {
            # Verbs
            'गर्दै': 'गर', 'गर्छ': 'गर', 'गर्नु': 'गर', 'गरे': 'गर', 'गर्यो': 'गर', 'गर्न': 'गर',
            'भन्ने': 'भन', 'भन्छ': 'भन', 'भन्यो': 'भन', 'भन्न': 'भन', 'भन्दै': 'भन',
            'हुने': 'हु', 'हुन्छ': 'हु', 'हुन': 'हु', 'हुँदा': 'हु', 'भएको': 'हु',
            'रहेछ': 'रह', 'रह्यो': 'रह',
            'आउँदै': 'आ', 'आउँछ': 'आ', 'आयो': 'आ',
            'जान्छ': 'जा', 'जादै': 'जा', 'गयो': 'जा',
            'लिन्छ': 'लि', 'लियो': 'लि',
            'दिन्छ': 'दि', 'दियो': 'दि',
            # Nouns/Adjectives
            'मानिसहरू': 'मानिस', 'कुराहरू': 'कुरा', 'साथीहरू': 'साथी',
            'छोराछोरी': 'छोरा', 'बच्चाबच्ची': 'बच्चा',
            'सानो': 'सान', 'ठूलो': 'ठूल'
        }

        # Words to never stem
        self.do_not_stem = {
            # Proper nouns
            'नेपाल', 'काठमाडौं', 'पोखरा', 'विराटनगर', 'ललितपुर', 'भक्तपुर', 'हिमाल', 'एवरेस्ट', 'कोशी',
            # Political Parties & Terms
            'नेकपा', 'माओवादी', 'कांग्रेस', 'एमाले', 'प्रचण्ड', 'पुष्पकमल', 'दाहाल', 'सचिवालय', 'प्रदेश',
            # Common nouns (important context)
            'सरकार', 'प्रधानमन्त्री', 'मन्त्री', 'बैंक', 'अदालत', 'संसद', 'बैठक', 'निर्णय', 'अभियान', 'भ्रष्टाचार',
            # Time-related
            'आज', 'हिजो', 'भोलि', 'बिहान', 'बेलुका', 'सप्ताह', 'महिना', 'वर्ष',
            # Loanwords
            'टेलिफोन', 'इन्टरनेट', 'मोबाइल', 'कम्प्युटर', 'डक्टर', 'इन्जिनियर'
        }

    def preprocess_text(self, text):
        """Main method to preprocess a single text string."""
        if not isinstance(text, str):
            return ""

        # 2. Remove punctuation/numbers/english (keep Nepali characters and spaces)
        text = re.sub(r'[a-zA-Z0-9\!\%\[\]\"\।\–\-\—\’\‘\’\’\#\$\%\&\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\@\^\_\`\{\|\}\~]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()

        # 3. Tokenize
        tokens = text.split()

        # 4. Remove Stopwords
        tokens = [t for t in tokens if t not in self.stop_words and len(t) > 1]

        # 5. Apply Stemming
        processed_tokens = []
        for token in tokens:
            if token in self.do_not_stem:
                processed_tokens.append(token)
                continue

            # Check irregular forms first
            stemmed_token = self.irregular_forms.get(token, None)
            if stemmed_token is not None:
                processed_tokens.append(stemmed_token)
                continue

            # Apply suffix patterns
            for pattern, replacement in self.suffix_patterns:
                orig_token = token
                token = re.sub(pattern, replacement, token)
                if token != orig_token: # Stop after first match
                    break

            processed_tokens.append(token)

        return processed_tokens
```

```python
# Initialize the preprocessor
preprocessor = NepaliTextPreprocessor()
```

```python
# Apply preprocessing to the dataframe
print("Preprocessing text...")
df['processed_tokens'] = df['content'].apply(preprocessor.preprocess_text)
```

```python
# Create a text version for debugging
df['processed_text'] = df['processed_tokens'].apply(lambda x: ' '.join(x))
```

```python
# --------------------- 4. TOPIC MODELING WITH LDA ---------------------
print("Creating dictionary and corpus...")
# Create a Dictionary and Corpus
dictionary = corpora.Dictionary(df['processed_tokens'])
# Filter extremes (remove very rare and very common words)
dictionary.filter_extremes(no_below=5, no_above=0.5)
corpus = [dictionary.doc2bow(tokens) for tokens in df['processed_tokens']]
```

```python
# Train the LDA Model
print("Training LDA model...")
num_topics = 10
lda_model = models.LdaModel(corpus=corpus,
                           id2word=dictionary,
                           num_topics=num_topics,
                           random_state=42,
                           passes=10,
                           alpha='auto',
                           per_word_topics=True)
```

```python
# Print the topics
print("\nDiscovered Topics:")
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}\n")
```

```python
# --------------------- 5. TEMPORAL TREND ANALYSIS ---------------------
print("Analyzing temporal trends...")
# Get the dominant topic for each document
def get_document_topic(bow):
    topic_dist = lda_model.get_document_topics(bow, minimum_probability=0)
    return max(topic_dist, key=lambda x: x[1])[0] # Return the topic index with highest probability

df['dominant_topic'] = [get_document_topic(doc) for doc in corpus]
```

```python
# Set the date as index and resample by month to see topic trends
df_time_series = df.set_index('date')
topic_trends = df_time_series.groupby([pd.Grouper(freq='M'), 'dominant_topic']).size().unstack(fill_value=0)
```

```python
# Plot the trends
plt.figure(figsize=(14, 8))
topic_trends.plot(kind='area', stacked=True, colormap='tab20')
plt.title('Trend of Dominant Topics in Nepali News Over Time')
plt.ylabel('Number of Articles')
plt.xlabel('Date')
plt.legend(title='Topic', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# --------------------- 6. VISUALIZATION (pyLDAvis) ---------------------
print("Preparing visualization...")
# Prepare the visualization
vis_data = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis_data)
# To save: pyLDAvis.save_html(vis_data, 'nepali_news_lda.html')
```
